#Introduction - Problem definition


The simple pendulum is a classic example of a physical system that exhibits oscillatory motion. The utility of the pendulum lies in its potential applications as a tool that can be used to study a wide range of physical phenomena, from timekeeping to earthquake measurement.

In physics classes, we model the behavior of a simple pendulum by using the small angle approximation. This estimate, which works well for small initial angles, fails however at capturing the behavior of a real pendulum for larger initial angles. Indeed, a simple pendulum is a non-linear system, meaning that the relationship between the force acting on it and its motion is not linear, making it difficult to solve analytically. One solution is to use the Velocity-Verlet algorithm and Euler's method to solve ODEs.

The main goal of this project is to evaluate the impact and precision of diverse mathematical models on the dynamic characteristics of a simple pendulum. We wish to discern the degree to which these models approximate or mimic the authentic behavior of the pendulum, while also considering the complexity of its motion and the equations governing its dynamics. To achieve this objective, we will compare the results obtained from various numerical models to simulations of realistic pendulums, and investigate the effects of parameters such as initial angle or time step duration on the motion of the pendulum. By doing so, we hope to gain a better understanding of the physics of the simple pendulum and its practical applications.

---
The following is a review of equations concerning the behavior of pendulums that employ the Small Angle Approximation (SA), Euler's Method and the Verlet Algorithm:

1. To implement **the Small Angle Approximation (SA)**, we assumed that the angle of displacement of the pendulum was small enough such that the sine of the angle is approximately equal to the angle itself. Using this approximation, we derived an analytical solution for the motion of the pendulum.

*   The angular frequency of a SA pendulum ($ùõö$) is given by:  $ùõö = \sqrt{g/L}$
*   Its angular acceleration ($\alpha$) is given by: $ Œ± = -g/L*Œ∏$
*   Its angular velocity (v) is given by:  $v = -A ùõö \sin(ùüÇ\Delta t)$

*   Its position ($Œ∏$) is given by: $ Œ∏ = A \cos(ùõö \Delta t)$



Therefore, by calculatung the angular frequency of a SA pendulum, we can get the pendulum's position and velocity at every time step. Then, we can also get its acceleration using the pendulum's position at each moment.

---
2. **Euler's method** is a numerical method used to approximate the solution of differential equations. It is a simple and widely used method, and shares some characteristics with the Verlet Algorithm. However, it is not as accurate nor stable as Verlet's. The formulas that describe Euler's method are as follows:

*   Its angular acceleration is given by: $Œ±_i = -g / l  *\sin(Œ∏_i)$
*   Its angular velocity is given by: $v_{i+1} = v_i + \alpha_i \Delta t$
*   Its position is given by: $Œ∏_{i+1} = Œ∏_i + v_i \Delta t + \frac{1}{2} Œ±_i  \Delta t^2 $

*   Its angular frequency is given by: $ ùõö = 2œÄ/T$ where $T$ is its period .

---
3. **The Verlet algorithm** is a numerical method used to solve differential equations of motion. It is particularly useful for simulating the motion of a pendulum because it is a second-order algorithm that conserves energy.
The equations that describe the Verlet Algorithm are as follows:

*   Its angular acceleration is given by: $Œ±_i = -g / l * \sin(Œ∏_i)$
*   Its angular velocity is given by: $v_{i+1} = v_i + \frac{\alpha_i+\alpha_{i+1}}{2} \Delta t $
*   Its position is given by: $Œ∏_{i+1} = Œ∏_i + v_i \Delta t + \frac{1}{2} Œ±_i  \Delta t^2 $

*   Its angular frequency is given by: $ ùõö = 2œÄ/T$ where $T$ is its period .







# Model and numerical methods


---
**The numerical method**

To simulate the motion of the pendulum, we create 4 lists with initial values. Then, we use the current values in the lists to calculate the updated time, position, velocity, and acceleration. These new values are then added to the respective lists. We continue these calculations until one complete oscillation of the pendulum is completed.


For the simulation of all 3 types of pendulums, we will create a Parent Class called **Pendulum**.

This class will contain:
1. Initial attributes and constants
2. Initialization of lists
3. Update methods
4. Other useful methods
5. Simulation loop

---
1. **Initial attributes and constants**

All 3 pendulums share the same attributes and constants:

*   Length (l)
*   Gravitational constant (g)
*   Max duration of simulation (T_MAX)
*   Length of time step (DT)
*   Initial angle (theta_0)
*   Initial angular velocity (v_0)

In the parent Class Pendulum, we obtain the attributes from keyword arguments. Other constants, such as g or T_MAX will be pre-defined class variables.

We also modify these attributes to fit our needs.

* Since the initial angle is given in degrees, we convert this angle into radians.

We also get other values from our attributes, which will used throughout the simulation. These values include:

* alpha_0: The initial angular acceleration, used as the first value in the alpha list
* w_SA: The angular frequency of a SA pendulum, used as a constant in the SA equations.
* STEPS: The number of STEPS of the simulation, obtained from the following formula: ```STEPS = T_MAX/DT```. It will serve as the number of loops in the simulate method.

---
2. **Initialization of lists**

We initialize the 4 lists and their respective inital values.
```
t = [0,]
theta = [theta_0,]
v= [v_0,]
alpha = [alpha_0, ]
```
Note that alpha_0 for the SA pendulum is different from Verlet's. Therefore, we calculate alpha_0 for SA then replace that value in the alpha list.

---
3. **Update methods**

We define "method" as being a function inside a class. For the Pendulum parent class, we will create 4 methods:
```
update_t(i)
update_theta(i)
update_v(i)
update_alpha(i)
```




Every method intakes the latest values inside the equation then appends this new value to its respective list. For example, update_t takes the latest value of time, then adds DT. This new value of time is then appended to the time list. In the next iteration, update_t will take this latest time value then add DT again. This, in essence, is our numerical method; we append values to a list based on the current values of the lists.

The parent class Pendulum will only contain the update methods for the Verlet Algorithm. The reason for this choice is because Euler and SA share some methods with Verlet, such as the time update method in both cases, or the acceleration update method in Euler's case. Therefore, by using reusing methods, reduce redundancy and thus increase the efficiency of our program.

Methods which are not shared with Verlet will be ascribed to their respective child class. For example, Euler's position update method is not the same as Verlet's. As such, we create a child class of Pendulum, called Euler, and create an update_theta method that has the equation for the position of a Euler pendulum. This will replace Verlet's update_theta with Euler's update_theta.

For each child class, we will list the methods that will replace those of the parent class Verlet:


```
class Euler(Pendulum):
  update_theta(i)
  update_v(i)
```

```
class SmallAngle(Pendulum):
  update_theta(i)
  update_v(i)
  update_alpha(i)
```
---

4. **Other methods**

Recall that the angular frequency of a Verlet and Euler pendulum is obtained from the following formula: ùõö = 2œÄ/T, where T is the duration of one oscillation.

We will create a method called ```w``` to obtain the angular frequency. Since update_t continuously appends time values to a list until one oscillation is complete, we are thus able to acquire T by retrieving the final value of the time list. We then return the value of the angular frequency of the pendulum after applying the formula listed above.

---
5. **Simulate method**

Now that we have described the initializition of lists and their corresponding update methods, we need a way to continuously append values to those lists.

We define ```stimulate(DT)``` as our simulation method. From the value of STEPS calculated, we iterate the update methods over the number of STEPS. For every iteration, the update methods will be called in this order:

```
update_t(i) #1
update_theta(i) #2
update_alpha(i) #3
update_v(i) #4
```
The reason for this specific order is because updating the velocity equation for the Verlet algorithm requires the average acceleration, which we get by calculating the average of the current acceleration and its new value, where the last variable requires the new theta. This also enables the Verlet algorithm to be reversible.

To make the simulation only simulate one oscilliation of the pendulum, we add a break statement when all of these conditions are met:

* There is a change in the sign of the velocity: this means that the pendulum is is reversing its motion after reaching its highest point or amplitude.
* The angular acceleration is negative: starting with the base assumption that the initial theta of the pendulum is a positive angle, a negative angular acceleration means that pendulum currently has positive angles.

By using these two conditions in conjunction, we guarantee that the simulation will stop when the pendulum is on the same side as the initial starting theta and that it is in the process of reversing its direction (the position is at an amplitude).

Furthermore, since the objective of this program is to ascertain the level of precision of several types of number methods, and some methods become inaccurate with certain initial conditions, we also add a break statement when the current theta surpasses 200% of the initial theta.

---
**Running the simulation**

To create our graphs that will serve our analysis, we set our initial conditions then create an object of the desired pendulum. For example, to create an object for the Verlet pendulum, we will use the class Pendulum, with all of the inputted initial conditions, and to create an object for the SA pendulum, we use the class SmallAngle. Then, by calling the method simulate on each object, we can therefore retrieve all the 4 lists that will be used to plot our graphs. Now, let's suppose that we want to investigate how different initial conditions affect the behavior of a pendulum and create a graph to visualize this. In that case, we can loop through a range of values and create a new pendulum object at each iteration with its initial conditions. Overall, this highlights the strengths of using object-oriented programming, as we can reduce redundancy and increase the efficiency of our code. Since we have already defined a blueprint for creating pendulum objects, we can create new instances quickly and easily.

###Code for Class Pendulum




````python
'''import statements'''

import math
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (9,6)

import numpy as np

'''class is below'''

class Pendulum:

#global instances for all the methods in the class

  T_MAX = 10.0
  g = 9.81

  def __init__(self,**kwargs):

    '''unpacking and calculating variables'''

    # Unpacks initial angle, checks if 0 or 90, then converts to radians

    theta_0_deg = abs(kwargs.get("theta_0_deg"))

    if theta_0_deg > 90:
      raise TypeError("angle should not exceed 90 degrees")
    if theta_0_deg == 0:
      raise TypeError("initial angle should not be 0")

    self.theta_0 = math.radians(theta_0_deg)

    # Unpacks initial angular velocity and length of the string

    self.v_0 = kwargs.get("v_0")

    self.l = kwargs.get("length")

    #Calculating initial angular acceleration for Pendulum (not for SA)

    self.alpha_0 = -1 * self.g / self.l * math.sin(self.theta_0)

    #Calculating angular frequency for SA (not for normal Pendulum)

    self.w_SA = math.sqrt(self.g/self.l)

    '''creating initial conditions for 3 all simulations'''

    # Initial conditions for Verlet pendulum
    self.t = [0,]
    self.theta = [self.theta_0,]
    self.v= [self.v_0,]
    self.alpha = [self.alpha_0, ]


    '''Update Methods for Verlet pendulum'''

  def update_t(self,i):
    t = self.t[i]+self.DT
    self.t.append(t)
    return

  def update_theta(self,i):
    theta = self.theta[i]+self.v[i]* self.DT + 0.5 * self.alpha[i]* self.DT**2
    self.theta.append(theta)
    return

  def update_v(self,i):
    v = self.v[i]+0.5*(self.alpha[i]+self.alpha[i+1])*self.DT
    self.v.append(v)
    return

  def update_alpha(self,i):
    alpha = -1 * self.g / self.l * math.sin(self.theta[i+1])
    self.alpha.append(alpha)
    return

  '''calculation of angular frequency'''

  def w(self):
    t_final = self.t[-1]
    w = (2*math.pi)/t_final
    return w


  '''Simulating the pendulum'''

  def simulate(self,DT):
    self.DT = DT
    STEPS = int(self.T_MAX/DT)
    for i in range(STEPS):
      self.update_t(i)
      self.update_theta(i)
      self.update_alpha(i)
      self.update_v(i)
      if abs(self.theta[i]) > self.theta_0*2: #checks if theta surpasses 200% of the initial angle
        self.theta.pop()
        self.t.pop()
        break
      if self.v[i-1]>0 and self.v[i]<0 and self.alpha[i] <0: #stop after one period
        break

    return self.t,self.theta,self.v,self.alpha

'''==============================================================='''
'''Euler child class'''

class Euler(Pendulum):

  def __init__(self, **kwargs):
    super().__init__(**kwargs)

  def update_theta(self,i):
    theta = self.theta[i]+self.v[i]* self.DT
    self.theta.append(theta)
    return

  def update_v(self,i):
    v= self.v[i]+self.alpha[i]*self.DT
    self.v.append(v)
    return

'''Small Angle child class'''

class SmallAngle(Pendulum):

  def __init__(self, **kwargs): #since alpha_0 is different for Small angle, we have to call the init() from parent, replace alpha_0, then create the new alpha list using alpha_0
    super().__init__(**kwargs)

    self.alpha_0 = -1*self.g/self.l* self.theta_0
    self.alpha = [self.alpha_0, ]

  def update_theta(self,i):
    theta = self.theta_0 * math.cos(self.w_SA*self.t[i])
    self.theta.append(theta)
    return

  def update_v(self,i):
    v= -1 * self.theta_0  * self.w_SA * math.sin(self.w_SA*self.t[i])
    self.v.append(v)
    return

  def update_alpha(self,i):
    alpha = -1*self.g/self.l*self.theta[i+1]
    self.alpha.append(alpha)
    return
````

###Initial conditions and initializing object

````python
'''Set initial conditions here valid for all cells below'''

length = 10
theta_0_deg = 90 #in degrees
v_0 = 0
DT = 0.01
'''Creating the object P'''

p = Pendulum(
    theta_0_deg = theta_0_deg,
    v_0 = v_0,
    length = length,
    )

p_euler = Euler(theta_0_deg = theta_0_deg,
    v_0 = v_0,
    length = length,)

p_SA = SmallAngle(theta_0_deg = theta_0_deg,
    v_0 = v_0,
    length = length,)

'''getting all data from methods'''

t_verlet,theta_verlet,v_verlet,alpha_verlet=p.simulate(DT)
t_euler,theta_euler,v_euler,alpha_euler=p_euler.simulate(DT)
t_SA,theta_SA,v_SA,alpha_SA=p_SA.simulate(DT)

w = p.w()
w_SA = p_SA.w_SA
````

````python
import matplotlib.pyplot as plt
import numpy as np

'''these values will apply for the following simulation'''

length = 10
theta_0_deg = 90 #in degrees
v_0 = 0
max = 0
min = -3
log_min = 10**min
DT_ARRAY = np.logspace(max, min, 30)

diff_average_list = []

p_min = Pendulum(
      theta_0_deg = theta_0_deg,
      v_0 = v_0,
      length = length,
      )

t_min,theta_min,v_min,alpha_min = p_min.simulate(log_min)

for DT in DT_ARRAY:

  diff_list = []

  p = Pendulum(
      theta_0_deg = theta_0_deg,
      v_0 = v_0,
      length = length,
      )

  t_verlet,theta_verlet,v_verlet,alpha_verlet=p.simulate(DT)

  for i in range(len(theta_verlet)):
    diff = abs(theta_verlet[i]-theta_min[i])
    diff_list.append(diff)

  average = sum(diff_list)/len(diff_list)
  diff_average_list.append(average)

````

````python
import matplotlib.pyplot as plt
import numpy as np

'''these values will apply for the 2 following simulations'''

length = 10
theta_0_deg = 90 #in degrees
v_0 = 0

DT_ARRAY=np.logspace(-1,-3,50)

w_list =[]
w_list_euler = []

for DT in DT_ARRAY:
  p = Pendulum(
      theta_0_deg = theta_0_deg,
      v_0 = v_0,
      length = length,
      )

  p_euler = Euler(
      theta_0_deg = theta_0_deg,
      v_0 = v_0,
      length = length,
      )

  p.simulate(DT)
  p_euler.simulate(DT)

  w_list.append(p.w())
  w_list_euler.append(p_euler.w())


plt.plot(DT_ARRAY,w_list,label=f"Verlet")
plt.plot(DT_ARRAY,w_list_euler,label=f"Euler")

plt.xlabel("Time Step $\Delta t$ (s)")
plt.ylabel("Angular Frequency (Hz)")
plt.legend()
plt.title(f"Angular frequency vs $\Delta T$")
````
